# Part 2
## Data Dictionary

id - Patient unique code, Identifier, Valid range is any number

age - Age, Continuous, 0–120 years

bp - Diastolic blood pressure in mmHg, Continuous, Should be within 40–130 mmHg

sg - Urine specific gravity (Weight in comparison to distilled water), Categorical-Ordinal, Range is 1.001–1.035

al - Albumin in urine, Categorical-Ordinal, higher = worse kidney function

su - Sugar in urine, Categorical-Ordinal, 0–5

rbc - Red blood cells in urine, Binary, normal/abnormal

pc - Pus cells in urine, Binary, normal/abnormal

pcc - Pus cell clumps in urine, Binary, present/notpresent

ba - Bacteria in urine, Binary, present/notpresent

bgr - Blood glucose random in mg/dL, Continuous, 70–500 mg/dL

bu - Blood urea in mg/dL, Continuous, 7–20 mg/dL

sc - Serum creatinine in mg/dL, Continuous, 0.5–15 mg/dL

sod - Serum sodium in mEq/L, Continuous, 120–160 mEq/L

pot - Serum potassium in mEq/L, Continuous, 2.5–6.5 mEq/L

hemo - Hemoglobin in g/dL, Continuous, 3.5–20 g/dL

pcv - Packed cell volume in % (Proportion of blood that is made up of cells), Continuous, 20–60%

wc - White blood cell count in cells/mm^3, Continuous, 3000–20000 cells/mm^3

rc - Red blood cell count in millions/mm^3, Continuous, 2.0–7.0 millions/mm^3

htn - Hypertension, Binary, yes/no

dm - Diabetes mellitus, Binary, yes/no

cad - Coronary artery disease, Binary, yes/no

appet - Appetite, Binary, good/poor

pe - Pedal edema (Foot swelling), Binary, yes/no

classification - Chronic Kidney Disease diagnosis, Binary, ckd / notckd
